### 종목 변경사항 적용

```python
def renameTickers(df):
  # String > Number
  df['수량'] = df['수량'].astype('int32')
  df['거래금액'] = df['거래금액'].astype('float')
  df['외화거래금액'] = df['외화거래금액'].astype('float')
  df['단가/환율'] = df['단가/환율'].astype('float')

  # 종목 변경사항 정리
  renamed = {
      'KINDEX 미국나스닥100' : 'ACE 미국나스닥100',
      '유니티 소프트웨어'      : '유니티소프트웨어',
      '페이스북'             : '메타 플랫폼스',
      '라이어트 블록체인'      : '라이엇 플랫폼즈',
      '어도비 시스템즈'        : '어도비',
      '아마존 닷컴'           : '아마존닷컴'
  }
  df['종목명'] = df['종목명'].apply(lambda x:renamed[x] if x in renamed.keys() else x)

  return df

df = thistory = renameTickers(thistory)
```

### Preprocessing

```python
SELLTYPE = ['매도','미국(NASDAQ)주식매도','미국(NYSE)주식매도','액면분할출고','액면병합출고','합병출고','분할출고','신주인수권출고','외화RP매도']

def calcTradeAmount(df, selltype):
  df['buysell'] = df['수량'] * np.where(df.거래내용.isin(selltype), -1, 1)
  df['buysell'] = np.where(df.거래내용.isin(['청약']), 0, df['buysell'])

  # KRW : 매수일때 -, 매도일때 +가 되는 부분임.

  df['KRW'] = df['거래금액'] * np.where((df.거래내용.isin(selltype)) | ((df.거래내용.str.contains('입금')) | (df.거래내용=='청약입고')), 1, -1)
  # df['KRW'] = df['단가/환율'] * np.where((df.거래내용.isin(selltype)) | (df.거래내용=='청약입고'), 1, -1)
  # df['KRW'] = np.where(df.거래내용.isin(['청약']), 0, df['거래금액'])

  df['USD'] = df['외화거래금액'] * np.where((df.거래내용.isin(selltype)) | (df.거래내용.str.contains('입금')), 1, -1)
  # df['USD'] = np.where(df.거래내용.isin(['청약']), 0, df['buysell'])

  # 0으로 reset 해야 하는 경우 
  df['KRW'] = np.where(df.통화명=='USD', 0, df['KRW'])
  df['USD'] = np.where(df.통화명=='KRW', 0, df['USD'])

  df['KRW'] = df['KRW'].astype('int')

  # 외화매수 : KRW = -, USD = - : KRW = -, USD = +로 변경
  df['USD'] = np.where(df.거래내용=='외화매수',0 - df['USD'], df['USD'])
  
  # thistory.loc[thistory['통화명']==''].종목명.unique()     # 통화명이 없는 것은 모두 KRW, 채권, 펀드 등임
  df['통화명'] = np.where(df['통화명']=='', 'KRW', df['통화명'])
  
  return df

thistory = calcTradeAmount(thistory, SELLTYPE)
```

## DATA 정리를 위한 함수들
- itemInfo(df, account:str, ticker:str) : thistory, idx의 계좌내의 ticker종목만 뽑아 매수수량누적, 매수평단가를 기록하여 이력을 dataframe으로 반환
- accInfo(df, account:str) : thistory, idx계좌의 종목별 dataframe을 준비하여 dict로 반환함 ticker : [최종보유수량, ticker이력의 dataframe]
- balanceTickerInfo(df, account:str) : thistory, idx의 계좌에 남아있는 종목들의 거래내역들 정리사항을 반환
- assembleStocks(tickerDict:dict) : 종목별로 df구성한 dict에서 전체 df를 생성
- balanceAccount(df, account:str) : 해당 계좌의 거래내역, 잔고를 산출
- balanceClass(dfhist, cat:str, catfc=None): Class 즉, 계좌분류(Family,YI,JW,MK)별 거래내역 및 잔고 확인


```python
def itemInfo(df, account:str, ticker:str):
  '''
  itemInfo : thistory, idx의 계좌내의 ticker종목만 뽑아 매수수량누적, 매수평단가를 기록하여 이력을 dataframe으로 반환
  '''
  item = df.loc[(df['계좌번호']==account)&(df['종목명']==ticker)].sort_index()
  item['ACC'] = item['buysell']
  item['ACC'] = item['ACC'].expanding(2).sum()
  item.loc[item.index[0], 'ACC'] = item.loc[item.index[0], 'buysell']
  
  item['AVG_KRW'] = np.where((item['buysell']!=0)&(item['KRW']<=0), item['KRW'].astype('float'), 0.0)
  item['AVG_USD'] = np.where((item['buysell']!=0)&(item['USD']<=0), item['USD'], 0.0)

  item['AVG_KRW'] = item['AVG_KRW'] / item['수량']
  item['AVG_USD'] = item['AVG_USD'] / item['수량']

  idx = 0 if item.iloc[0, -2] != 0 else 1
  item['AMT_PRICE'] = item['KRW'] if idx==0 else item['USD']

  OP = 1
  AVG_ = -3+idx
  TRADE = -6+idx
  AMT_PRICE = -1
  BUYSELL = -7
  ACC = -4

  item['AMT_PRICE'] = item['AMT_PRICE'].expanding(2).sum()     # 해당 종목의 누적치
  item.iloc[0, -1] = item.iloc[0, TRADE]

  # 평단가 계산
  for i in range(1, item.shape[0]):
    if item.iloc[i, BUYSELL]==0:     # buysell
      item.iloc[i, -3] = np.abs(item.iloc[i-1, -3])     # AVG_KRW, AVG_USD
      item.iloc[i, -3+1] = np.abs(item.iloc[i-1, -3+1])
      # 액면분할입고/액면분할출고 처리 
      continue
    if item.iloc[i, BUYSELL] < 0:
      item.iloc[i, AVG_] = np.abs(item.iloc[i-1, AVG_])     # AVG_KRW, AVG_USD
    else:
      # 거래내용 == 분할입고면 분할입고 수량만 참고하여 평단가 계산 ... 아래 코드로 대체해 봄
      # nStock = item.iloc[i, 3] if item.iloc[i, 1] == '분할입고' else item.iloc[i, ACC]
      nStock = item.iloc[i, ACC]
      if item.iloc[i-1, AVG_] != 0 and item.iloc[i-1, ACC] != 0:
        item.iloc[i, AVG_] = (np.abs(item.iloc[i-1, ACC] * item.iloc[i-1, AVG_]) + np.abs(item.iloc[i, TRADE])) / nStock     # 평단가와 거래금액은 부호가 반대이므로 -
      else:
        if i > 2 and '분할입고' in item.iloc[i, OP]:     # 액면분할입고, 분할입고
          # print(f'i = {i}, OP={item.iloc[i, OP]}')
          # 마지막 AVG_KRW / AVG_USD 찾기 -> ACC * AVG_KRW or AVG_USD / 액면분할입고 수량 = 새로운 AVG_   ; 수량이 떨어지지 않는 경우가 있나 모르겠음
          srch = i - 1
          while(srch >= 0):
            if item.iloc[srch, ACC] > 0 and item.iloc[srch, AVG_] > 0:
              break
            srch = srch - 1
          if srch >= 0:   # FOUND !
            item.iloc[i, AVG_] = item.iloc[srch, ACC] * item.iloc[srch, AVG_] / item.iloc[i, ACC]
          else :          # NOT FOUND !
            pass

    if item.iloc[i, ACC]==0:
      item.iloc[i, AVG_] = 0
      trade = 0

  item['AVG_KRW'] = np.abs(item['AVG_KRW']).fillna(0)
  item['AVG_USD'] = np.abs(item['AVG_USD']).fillna(0)

  # 채권은 왜 0이 되는가? 이거 확인 후 풀어야 할 듯
  if False:
    item['AVG_KRW'] = item['AVG_KRW'].astype('int')
    item['AVG_USD'] = item['AVG_USD'].astype('float')

  # 삼성전자때문에 넣어 봄...
  item['AVG_KRW'] = np.where((item['거래내용']=='매수')&(item['AVG_KRW']==0), item['단가/환율'].astype('float'), item['AVG_KRW'])

  return item

def accInfo(df, account:str):
  '''
  accInfo : thistory, idx계좌의 종목별 dataframe을 준비하여 dict로 반환함 ticker : [최종보유수량, ticker이력의 dataframe]
  '''
  acc = df.loc[(df['계좌번호']==account)].sort_index()
  tickers = acc['종목명'].unique()
  items = {}
  for t in tickers:
    i = itemInfo(df, account, t)
    nStock = i.iloc[-1, -4] if i.shape[0] > 0 else 0
    items[t] = [nStock, i]

  return items

def balanceTickerInfo(df, account:str):
  '''
  balanceTickerInfo : thistory, idx의 계좌에 남아있는 종목들의 거래내역들 정리사항을 반환
  '''
  accN = accInfo(df, account)

  acc_c = { k:accN[k] for k in accN.keys() if accN[k][0] != 0}
  _dflist = []
  for k in acc_c.keys():
    _df = pd.DataFrame(acc_c[k][1].sort_index().iloc[-1, [2, -3, -2, -1]]).T.reset_index().iloc[:, 1:]
    _df['ACC'] = acc_c[k][0]
    _dflist = _dflist + [_df]
  acc_cdf = pd.concat(_dflist)

  for i in range(1, 3):
    acc_cdf.iloc[:, i] = np.abs(acc_cdf.iloc[:, i])

  acc_cdf['CURRENCY'] = np.where(acc_cdf.iloc[:, -2] != 0, 'USD', 'KRW')
  
  return accN, acc_cdf

def assembleStocks(tickerDict:dict):
  '''
    assembleStocks : 종목별로 df구성한 dict에서 전체 df를 생성
  '''
  dfa_list = []
  for k in tickerDict.keys():
    dfa_list = dfa_list + [tickerDict[k][1]]
  dfa = pd.concat(dfa_list).sort_index()

  return dfa

def balanceAccount(df, account:str):
  '''
  해당 계좌의 거래내역, 잔고를 산출
  '''
  # ''은 thistory부터 있으니 나중에 확인해 볼 것!
  # 0번 계좌에 남아있는 종목들의 거래내역들 정리사항을 반환
  acc0, acc0c = balanceTickerInfo(thistory, account)
  # acc0c     # 실제 잔고, 평단가 확인해 보자

  # AVG_, AMT_ 를 포함하여 다시 해당 계좌 전체 데이타를 취합 : 예수금 처리 가능하지 않을까? (이전 데이타가 없어서... ???)
  df0 = assembleStocks(acc0)

  # 현재 보유중인 종목과 수량, 금액, 평균매수단가
  df0c = df0.groupby('종목명')[['buysell' ,'KRW', 'USD']].sum()
  df0cr = df0c.loc[df0c['buysell']!=0].abs().sort_values(['KRW', 'USD'], ascending=False)
  df0a = pd.DataFrame({'AVG_KRW':[acc0[x][1].iloc[-1, -3] for x in df0cr.index], 'AVG_USD':[acc0[x][1].iloc[-1, -2] for x in df0cr.index]}, index=df0cr.index)
  df0f = df0cr.merge(df0a, left_index=True, right_index=True, how='left')
  
  # 계좌별 현금잔고 관련
  df0['CASH_KRW'] = df0['KRW']
  df0['CASH_USD'] = df0['USD']
  # df['CASH_KRW'] = np.where((df.KRW<0)&(~df.거래내용.isin(['매수'])), 0 - df['CASH_KRW'], df['CASH_KRW'])   # 이미 -로 CASH에서 - 되도록 되어 있음
  # df['CASH_USD'] = np.where((df.USD<0)&(~df.거래내용.str.contains('매수')), 0 - df['CASH_USD'], df['CASH_USD'])   # 상동
  df0['CASH_KRW'] = df0['CASH_KRW'].expanding(2).sum()
  df0['CASH_USD'] = df0['CASH_USD'].expanding(2).sum()
  df0.iloc[0, -2] = df0.iloc[0, -8]
  df0.iloc[0, -1] = df0.iloc[0, -7]

  df0f = df0f.loc[:, ['buysell', 'AVG_KRW', 'AVG_USD']]
  df0f['매수금액(KRW)'] = df0f['buysell'] * df0f['AVG_KRW']
  df0f['매수금액(USD)'] = df0f['buysell'] * df0f['AVG_USD']

  return df0, df0f

def balanceClass(dfhist, cat, catfc=None):
  '''
  balanceClass : Class 즉, 계좌분류(Family,YI,JW,MK)별 거래내역 및 잔고 확인
  '''
  df_list = []
  dff_list = []
  df = dfhist.loc[dfhist.category==cat]
  acclist = df.계좌번호.unique()
  for acc in acclist:
    try:
      df, df_f = balanceAccount(dfhist, acc)
      df_list.append(df)
      dff_list.append(df_f)
    except:
      print(f'skip {acc} ...')

  df = pd.concat(df_list)
  df_f = pd.concat(dff_list)
  df_f1 = df_f.reset_index().groupby('종목명')['buysell'].sum()     # 전체 sum을 하면 안될 듯, average처리도 필요 (col : buysell	KRW	USD	AVG_KRW	AVG_USD)
  df_f2 = df_f.reset_index().groupby('종목명')['AVG_KRW','AVG_USD'].mean()
  df_f = pd.concat([df_f1, df_f2], axis=1)
  df_f['매수금액(KRW)'] = df_f['buysell'] * df_f['AVG_KRW']
  df_f['매수금액(USD)'] = df_f['buysell'] * df_f['AVG_USD']

  if catfc != None:
    df, df_f = catfc(df, df_f)
  return df, df_f
```

## 자산분류를 위한 함수
- categoriging(quant=False) : quant 여부를 참고하여 df와 df_f에 자산분류를 적용하는 함수를 반환

```python
def categoriging(quant=False):
  '''
  quant 여부를 참고하여 df와 df_f에 category를 적용하는 함수를 반환
  '''
  def _fc(df, df_f):
    if quant==False:
      dfcat = category2
    else:
      dfcat = category
    df2 = df.merge(right=dfcat, left_on='종목명', right_on='종목명')
    dff2 = df_f.reset_index()
    dff2 = dff2.merge(right=dfcat, left_on='종목명', right_on='종목명')

    return df2, dff2
  return _fc
```
